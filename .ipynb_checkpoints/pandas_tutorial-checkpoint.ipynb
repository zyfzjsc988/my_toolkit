## 创建 DataFrame

pandas 有它自己的 datetime 会直接变成index形式

```python
dates = pd.date_range('20130101', periods=6) 
dates
```

DataFrame 的格式是 pd.DataFrame(data,index=INDEX,columns=list)

- 一般没有指定index的话，就是自动生成的从0开始的序列

- 一般random data用 numpy的random。 np.random.randn(row,column)

```python
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
```

如果参数是一个 dict，每个 dict 的 value 会被转化成一个 Series

格式是，pd.DataFrame({'列名':列中的内容})   
- 列中的内容可以是list也可以是一个值，如果一个值，当行数超过1时会被广播。
- 最基础的应该是pd.Series 类型。 可以认为，DataFrame 是由 Series 组成的
- 增加一列 Series， s1 = pd.Series([1,2,3,4,5,6], index=pd.date_range('20130102', periods=6))， df['E'] = s1

```python
df2 = pd.DataFrame({ 'A' : 1.,
                    'B' : pd.Timestamp('20130102'),
                    'C' : pd.Series(1,index=list(range(4)),dtype='float32'),
                    'D' : np.array([3] * 4,dtype='int32'),
                    'E' : pd.Categorical(["test","train","test","train"]),
                    'F' : 'foo' })
df2
# df2.dtypes() #检查各列数据类型
```

## 选择

简易选择器
- 行 df[index:index] ,df[0:3]  或 df[index名:]
- 列 df['列名'], 也可以使用 df.列名

```python
df2[['B','C']]
df2[:1]
df['20130102':'20130104']
```

通过 label 选择  (loc)
- 选择多个元素时  df.loc['20130102':'20130104',['A','B']] [index:index,[列名]]。不输入列名时会输出整行，需要输出整列则需要df.loc[:,[列名]]
- 选择一个元素时  df.at[dates[0],'A']

通过整数下标选择 (iloc)
- df.iloc[3:5,0:2]
- df.iloc[[1,2,4],[0,2]]
- 单个元素 df.iloc[1,1]， df.iat[1,1]

布尔条件选择器
- df[df.A > 0] 取出特定列
- df[df > 0] 返回整个df，但是不符合的部分变成NAN
- df2[df2['E'].isin(['two','four'])]， isin函数是判断元素是否在某个集合内。



```python
df[df.A > 0]
df[df > 0]
```

## set reset Index, columns name

Reindexing 允许你改变某个轴的 index: 此功能仅仅是改变index而已，相当于扩展或者压缩轴中的数据。
- df.reindex(index=dates[0:4], columns=list(df.columns) + ['E'])
- 可以对index也可以对columns进行扩展，缺失的值会填充NAN，也可以使用fill_value=。
- 另一种形式 ： (labels, axis={'index', 'columns'}, ...)

```python
date_index2 = pd.date_range('20130101', periods=8) 
df
df.reindex(date_index2, method='ffill')
```

## Merge

concat: 横向或纵向拼接 
- pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=, names=, verify_integrity=False)
- axis=0 index为纵向，1 columns为横向
- verify_integrity 会检查重复键
- pd.concat([s1, s2], keys=['s1', 's2'], names=['Series name', 'Row ID']) 表示将每部分的数据用新的一级index标识，names是从index开始标的列名

merge: 可以与join等价。熟悉其中之一就可以。
- DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False)
- merge by columns:Key 是 columns name,  df3.merge(df,how='inner')
- merge by index: df3.merge(df.rename_axis('index'),how='inner',left_index=True,right_index=True)

Append: 增加一行
- df.append(s, ignore_index=True) ， s需要是Series

```python
df3= df.rename(columns={'A':'A1'}).rename_axis('index')
df3
df3.merge(df.rename_axis('index'),how='inner',left_index=True,right_index=True)
```

## reshape 改变 df 结构， 一般用于有categories的df

stack() 把 DataFrame 的列“压缩”到 index 里去
- df2.stack()

unstack() 恢复出列，默认把最后一个 index 解开
- df.unstack(index列序号)

Pivot Table 旋转: 仅限 categories
- pivot 是把原来的数据(values)作为新表的行(index)、列(columns)
- pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])

```python
# 准备数据集
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz','foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']]))
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
df2 = df[:4]
df2
stacked = df2.stack()
stacked
stacked.unstack()
stacked.unstack(0)
stacked.unstack().unstack()
```

```python
df = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
                   'B' : ['A', 'B', 'C'] * 4,
                   'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                   'D' : np.random.randn(12),
                   'E' : np.random.randn(12)})
df
pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])
```

## 时间序列
resample 功能：
- series.resample('L', how='bfill') 
- frequency : http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases
- how: sum, ffill, bfill,

```python
rng = pd.date_range('1/1/2012', periods=1000, freq='S')
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts.resample('T', how='ffill')
```

## 类别
将string list转成 categories，方便转换和summary
- df["raw_grade"].astype("category")
- df["grade"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])

```python
df = pd.DataFrame({"id":[1,2,3,4,5,6], "raw_grade":['a', 'b', 'b', 'a', 'a', 'e']})
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
df["grade"].cat.categories = ["very good", "good", "very bad"]
df["grade"] = df["grade"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])
df["grade"]
df.groupby("grade").size()
```
